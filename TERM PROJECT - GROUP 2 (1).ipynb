# Introduction


The Föppl–von Kármán equations are named after August Föppl and Theodore von Kármán are a set of nonlinear partial differential equations describing the large deflections of thin flat plates.With applications ranging from the design of submarine hulls to the mechanical properties of cell wall. In our problem statememnt we have considered Buckling of Uniformely compresssed Rectangular plate simply supported along two opposite sides perpendicular in the direction of compression and having various edge conditions along the other two sides.

<a id="prerequisites"></a>

# Prerequisites 

#### The first Föppl–von Kármán equation is
\begin{align*}
 N_x \frac{\partial^2 w}{\partial x^2}+ 2N_{xy} \frac{\partial^2 w}{\partial x \partial y}+N_y \frac{\partial^2 w}{\partial y^2}- D\nabla^4 w = -q
 \end{align*}
 $where$ \
  \begin{align*}
 \nabla^4 w = \frac{\partial^4 w}{\partial x^4}+\frac{\partial^4 w}{\partial^2 x \partial^2 y} + \frac{\partial^4 w}{\partial y^4} \
 \end{align*}


<a id="userep"></a>

# Useful representations

 $(1)D = \frac{Eh^3}{12(1-\nu^2)}$\
$(2) M_x = -D (\frac{\partial^2 w}{\partial x^2} + \nu \frac{\partial^2 w}{\partial y^2})$\
$(3) M_y = -D (\frac{\partial^2 w}{\partial y^2} + \nu \frac{\partial^2 w}{\partial x^2})$\
$(4) M_{xy} = -D (\frac{\partial^2 w}{\partial x \partial y} )$\
$(5) Q_x  =  \frac{\partial M_x}{\partial x} + \frac{\partial M_{xy}}{\partial y} $\
$(6) Q_y  =  \frac{\partial M_{xy}}{\partial x} + \frac{\partial M_y}{\partial y} $\
$(7) N_x = \frac{\partial F ^2}{\partial y^2}$\
$(8) N_y = \frac{\partial F ^2}{\partial x^2}$\
$(9) N_{xy} = -\frac{\partial F ^2}{\partial y \partial x}$\
 Where $F$ satisfies the equation,\
<font size="3.5">
    $\nabla^4F = Eh[(\frac{\partial w^2}{\partial x\partial y})^2 -\frac{\partial^2 w}{\partial x^2}\frac{\partial^2 w}{\partial y^2}]$\
 The above equation represents $2^{nd}$ Föppl–von Kármán equation.

# Problem statement

#### Here we are considering rectangular plate with dimensions $a$ and $b$ along $x$ and $y$ directions respectively. 
#### The Reactangular plate is simply supported  at the edges $x=0$ and $x=a$. 
#### The plate is subjected to a compressive load along the $x$ - direction and the magnitude of the load is $N_x$.
#### The special condition that we have considered is that at $y=0$ the plate is simply supported and at  $y=b$ it is free. 


<a id="soln"></a>

# Solution for Our Consideration:

The Föppl-von Kármán equation is given by: \
As we are loading in x- direction, only $N_x$ term will be there.\
Final Governing equation is,
<font size="3"> 
\begin{align*}
 D(\frac{\partial^4 w}{\partial x^4} + 2\frac{\partial^4 w}{\partial x^2 \partial y^2}+\frac{\partial^4 w}{\partial x^4}) = -N_x \frac{\partial^2 w}{\partial x^2}
 \end{align*}
     </font>
     
Since it is simply suported on the sides perpendicular to x-axis, boundary conditions on the sides $x=0$ and $x=a$ are:\
<font size="3"> $w=0$ </font> &emsp; and &emsp; <font size="5"> $\frac{\partial^2 w}{\partial x^2} + \nu \frac{\partial^2 w}{\partial y^2}$ </font> &emsp; &emsp; at &emsp; <font size="3.5"> $x=0$</font> &emsp; and &emsp; <font size="3.5">$x=a$ </font>

Similiarly the edge $y=0$ is also simply supported,\
<font size="3"> $w=0$ </font> &emsp; and &emsp; <font size="5"> $\frac{\partial^2 w}{\partial y^2} + \nu \frac{\partial^2 w}{\partial x^2}$ </font> &emsp; &emsp; at &emsp; <font size="3.5"> $y=0$</font>

Since the edge $y=b$ is free end Bending moment and shear force terms are zero, \
 <font size="5"> $\frac{\partial^2 w}{\partial y^2} + \nu \frac{\partial^2 w}{\partial x^2}$  </font> &emsp; and &emsp; <font size="5"> $\frac{\partial^3 w}{\partial y^3} + (2-\nu) \frac{\partial^3 w}{\partial x^2 \partial y}$ </font> &emsp; &emsp; at &emsp; <font size="3.5"> $y=b$</font>

Shear force term is ,\
<font size="5">$ Q_y - \frac{\partial M_{xy}}{\partial y}$</font>

```python
import sympy as sp
```

```python
x, y = sp.symbols('x, y')
a,b,Nx,D,w,m,nu=sp.symbols('a,b,N_x,D,w,m,nu',positive=True)
f=sp.Function('f')(y)
```

As discussed before simply suported on the sides perpendicular to $x$-axis, boundary conditions on the sides $x=0$ and $x=a$ are:\
<font size="3"> $w=0$ </font> &emsp; and &emsp; <font size="5"> $\frac{\partial^2 w}{\partial x^2} + \nu \frac{\partial^2 w}{\partial y^2}$ </font> &emsp; &emsp; for &emsp; <font size="3.5"> $x=0$</font> &emsp; and &emsp; <font size="3.5">$x=a$ </font>
 

 
   


So we are considering an expression for $w$ which satisfies the above two boundary conditions.\
    The term 
 <font size="4"> $sin(\frac{\pi mx}{a})$ </font>
     satisfies the above two boundary condtions. But $w$ is a function of both $x$,$y$.\
     $w$ must be defined in such a way that it should depend on $y$ also.

```python
from IPython.display import Math, Latex
```

```python
w=sp.sin(m*sp.pi*x/a)*f
display(Math(r'w = {}'.format(sp.latex(w))))
```

```python
def laplacian(f):
    return sp.diff(f,x,2)+sp.diff(f,y,2)
def biharmonic(f):
    return laplacian(laplacian(f))
```

```python
print('Governing differential equation : ')
eq=(sp.Eq(D*biharmonic(w) + Nx*sp.diff(w,(x,2)),0)).simplify()
display(eq)
```

From the above expression,for non trivial soultion we take the coefficient of <font size="4"> $sin(\frac{\pi mx}{a})$ </font> is equal to zero. \
Now the expression yields to \
<font size="5"> 
$\frac{d^4f}{dy^4} -\frac{2m^2\pi^2}{a^2} \frac{d^2f}{dy^2} + (\frac{m^4\pi^4}{a^4} - \frac{N_x}{D}\frac{m^2\pi^2}{a^2})f = 0$
    </font>

#### By solving the above equation


```python
print('On solving upon above differential equation, we get : ')
y_soln = sp.dsolve(eq).simplify()
display(y_soln)
```

### Useful substituions
Now lets \
<font size="4"> $ \alpha = \frac{\sqrt\pi\sqrt m \sqrt {\pi \sqrt D m - \sqrt N_x a}}{\sqrt[4]Da} = \sqrt {\frac{m^2 \pi^2}{a^2} + \sqrt{\frac{N_x}{D} \frac{m^2 \pi^2}{a^2}}}$ </font> \
<font size="4"> $\beta = \frac{\sqrt\pi\sqrt m \sqrt {\pi \sqrt D m + \sqrt N_x a}}{\sqrt[4]Da} = \sqrt {-\frac{m^2 \pi^2}{a^2} + \sqrt{\frac{N_x}{D} \frac{m^2 \pi^2}{a^2}}}$ </font>

```python
alpha, beta,A1,A2,A3,A4 = sp.symbols('alpha, beta,A_1,A_2,A_3,A_4,')
f1 = sp.Function('f')(y)
eqn = sp.Eq(f1,A1*sp.exp(-alpha*y) + A2*sp.exp(alpha*y) + A3*sp.cos(beta*y) + A4*sp.sin(beta*y))
display(eqn)
```

```python
print('Boundary conditions 1 : ')
bc1_lhs = eqn.rhs.subs(y,0)
bc1 = sp.Eq(bc1_lhs,0).simplify() # deflection at y=0 is 0
display(bc1)
```

```python
print('Boundary conditions 2 : ')
bc2_lhs = (sp.diff(eqn.rhs,(y,2)) + nu*sp.diff(eqn.rhs,(x,2))).subs(y,0)
bc2 = sp.Eq(bc2_lhs,0) # My at y=0 is 0
display(bc2)
```

From above two equations, \
$A_1 = -A_2-A_3---->(1)$ \
$A_1\alpha^2+A_2\alpha^2 - A_3\beta^2=0      ------> (2)$\
$(A_1 + A_2)\alpha^2-A_3\beta^2 =0$

from equation 1, substituting
$A_1 +A_2 =-A_3$  in equation 2 \
$-A_3\alpha^2-A_3\beta^2=0$\
$-A_3(\alpha^2+\beta^2)=0$ \
$A_3=0$


we can conclude that $A_3 =0$ and $A_1=-A_2$
then $f(y)$ yields to : \
$f(y) = A_1e^{-\alpha y}-A_1e^{\alpha y} + A_4 sin(\beta y)$\
Using $Sinh(x)$ = <font size="5"> $\frac{e^x-e^{-x}}{2}$  

On further simplifaction $f(y)$ can be written as,\
$f(y) = B_1 sinh(\alpha y) + B_2 sin(\beta y)$

```python
B1,B2 = sp.symbols('B_1,B_2')
f2 = sp.Function('f')(y)
eqn2 = sp.Eq(f2,B1*sp.sinh(alpha*y) + B2*sp.sin(beta*y))
display(eqn2)
```

```python
w_soln1 = ((eqn2.rhs)*(sp.sin(m*sp.pi*x/a))).simplify()
print('w =')
display(w_soln1)
```

```python
print('Boundary conditions 3 : ')
bc3_lhs = (sp.diff(w_soln1,(y,3)) + (2-nu)*sp.diff(sp.diff(w_soln1,(x,2)),y)).subs(y,b)
bc3 = sp.Eq(bc3_lhs,0).simplify() # dMns/ds at y=b is 0
display(bc3)
```

```python
print('Boundary conditions 4 : ')
bc4_lhs = (sp.diff(w_soln1,(y,2)) + nu*sp.diff(w_soln1,(x,2))).subs(y,b)
bc4 = sp.Eq(bc4_lhs,0).simplify() # My at y=b is 0
display(bc4)
```

When we try to solve the equations obtained by using boundary conditions is $3$ and $4$ , we get the arbitary constants will be zero. So again for non trivial solution, taking the determinant of coefficient matrix is zero

```python
term1 = bc3.lhs.expand().coeff(B1).simplify()/sp.sin(m*sp.pi*x/a)
display(term1)
```

```python
term2 = bc3.lhs.expand().coeff(B2).simplify()/sp.sin(m*sp.pi*x/a)
display(term2)
```

```python
term3 = bc4.lhs.expand().coeff(B1).simplify()/sp.sin(m*sp.pi*x/a)
display(term3)
```

```python
term4 = bc4.lhs.expand().coeff(B2).simplify()/sp.sin(m*sp.pi*x/a)
display(term4)
```

```python
M = sp.Matrix([[term1, term2], [term3, term4]])
display(M)
```

```python
eqn3= sp.Eq(M.det(),0).expand()
display(eqn3)
```

 On simplifying the above equation,

```python
eqn4 = sp.Eq(beta*(alpha**2-nu*m**2*sp.pi**2/a**2)**2*sp.tanh(alpha*b),alpha*(beta**2+nu*m**2*sp.pi**2/a**2)**2*sp.tan(beta*b))
display(eqn4)
```

Now substituting thevalues of $\alpha$ and $\beta$ in the above expresssion 

$ \alpha =\sqrt {\frac{m^2 \pi^2}{a^2} + \sqrt{\frac{N_x}{D} \frac{m^2 \pi^2}{a^2}}}$  \
$ \beta =\sqrt {-\frac{m^2 \pi^2}{a^2} + \sqrt{\frac{N_x}{D} \frac{m^2 \pi^2}{a^2}}}$

```python
eqn5 = eqn4.subs([(alpha,sp.sqrt((m**2)*(sp.pi**2)/(a**2) + sp.sqrt(Nx*(m**2)*(sp.pi**2)/(D*(a**2)))))])
eqn6 = eqn5.subs([(beta,sp.sqrt(-(m**2)*(sp.pi**2)/(a**2) + sp.sqrt(Nx*(m**2)*(sp.pi**2)/(D*(a**2)))))])
display(eqn6)
```

# Critical Buckling coefficient$(k_{cr})$:


From Theory of Elastic Stability by Timoshenko and Gere,\
The general expression for  $(N_x)_{cr}$ is,\
<font size="4"> $(N_x)_{cr} = k_{cr} \frac{\pi^2 D}{b^2h}$ </font>\
  where ,\
&emsp; <font size="4"> $k_{cr} = (0.456 + \frac{b^2}{a^2})$</font>

```python
a_by_b = np.arange(0.5,10,0.5)
b_by_a=1/a_by_b
k = (0.456 + (b_by_a)**2)
import tableprint as tp
data = np.column_stack((a_by_b,k))
headers = ['a/b', 'k']
tp.table(data, headers)
```

```python
fig = plt.figure(figsize=(10,10/1.8))
plt.plot(a_by_b,k,'k-');
plt.title(r'$Critical$ $buckling$ $ratio$ $(k_{cr})$ $vs$ $aspect$ $ratio$ $(\frac{a}{b})$',fontweight='bold',fontsize=25);
plt.xlabel(r'$a/b$',fontsize=16);
plt.ylabel(r'$0.456 + (b^2/a^2)$',fontsize=16);
```
